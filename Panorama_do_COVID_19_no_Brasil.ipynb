# Panorama do COVID-19 no Brasil

O **COVID-19** foi uma doença causada por um coronavírus recém-descoberto, que mudou completamente o curso da história recente. Embora a pandemia tenha sido um grande desafio global, hoje, com a situação mais controlada, conseguimos tirar lições valiosas sobre a saúde pública e a capacidade de adaptação de sociedades inteiras.

A transmissão do vírus ocorreu principalmente através de gotículas expelidas ao tossir ou espirrar, mas também se espalhou por superfícies contaminadas. A gravidade dos sintomas, como se sabia desde o começo, variava bastante: algumas pessoas nunca sentiram nada, outras tiveram apenas sintomas leves, mas uma parcela significativa desenvolveu quadros graves e precisou de cuidados intensivos.

Após anos de pesquisa e muitos estudos ao redor do mundo, hoje já sabemos que, no começo, a doença era rodeada de incertezas. Embora já existam protocolos mais claros e vacinas amplamente disponíveis, os primeiros meses da pandemia foram marcados pela falta de informações e pela necessidade de constante adaptação. Muitos aspectos do vírus ainda são estudados, mas, em termos gerais, conseguimos avançar bastante no entendimento de sua transmissão, sintomas e formas de tratamento.

Observa-se que, em média, 80% dos infectados apresentaram sintomas leves ou foram assintomáticos, recuperando-se sem maiores complicações. O restante, aproximadamente 20%, teve um curso mais grave da doença, com 15% necessitando de oxigênio suplementar e 5% requerendo ventilação mecânica, em ambientes hospitalares.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## **Obtenção dos Dados**

*Este dataset oferece uma visão detalhada da evolução da pandemia de COVID-19 em diferentes países e regiões, possibilitando análises sobre a propagação do vírus, suas taxas de mortalidade, os impactos das campanhas de vacinação, e as respostas de saúde pública a diferentes momentos da crise. Através das informações contidas nele, podemos observar como fatores como a densidade populacional, a infraestrutura de saúde, as políticas governamentais, e o comportamento social influenciam a dinâmica da pandemia.*

*Além disso, o dataset permite realizar comparações entre países, estados ou regiões, identificando tendências, como picos de infecção, taxas de mortalidade e os efeitos das campanhas de vacinação. Isso pode fornecer insights cruciais sobre quais medidas de saúde pública foram mais eficazes e onde houve lacunas no enfrentamento da crise. Contudo, a interpretação desses dados exige cautela, dada a complexidade e a variedade de variáveis envolvidas.*

*A análise de uma doença emergente, como o COVID-19, apresenta uma série de desafios únicos. Um dos maiores obstáculos é a falta de dados históricos robustos. Sendo uma doença nova, não há informações amplamente consolidadas ou experiências passadas que possam ser utilizadas para fazer previsões precisas. Além disso, a doença passa por mutações que alteram a forma como ela se transmite, como ela afeta as pessoas e como o sistema imunológico responde, o que complica ainda mais qualquer análise a longo prazo.*

*Outro desafio significativo é o tempo de resposta e a necessidade de aprendizado contínuo sobre a doença. Cientistas e profissionais de saúde estavam, e ainda estão, aprendendo sobre o vírus à medida que novos dados vão sendo coletados. As características do COVID-19, como sua taxa de mortalidade e transmissão, variam conforme novas variantes surgem, o que significa que modelos preditivos frequentemente precisam ser revisados e atualizados.*

## **Dicionário de Variáveis**
* iso_code - Código único do país ou região (padrão ISO 3166-1).
* continent - Continente onde o país ou região está localizado.
* location - Nome do país ou território.
* date - Data específica em que os dados foram registrados.
* total_cases - Número total de casos confirmados de COVID-19 até a data especificada.
* new_cases - Número de novos casos confirmados de COVID-19 em um dia específico.
* total_deaths - Número total de mortes causadas por COVID-19 até a data especificada.
* new_deaths - Número de novas mortes causadas por COVID-19 em um dia específico.
* total_cases_per_million - Número total de casos de COVID-19 por 1 milhão de pessoas.
* new_cases_per_million - Número de novos casos de COVID-19 por 1 milhão de pessoas.
* total_deaths_per_million - Número total de mortes por COVID-19 por 1 milhão de pessoas.
* new_deaths_per_million - Número de novas mortes por COVID-19 por 1 milhão de pessoas.
* total_tests - Número total de testes realizados para detectar a COVID-19.
* new_tests - Número de novos testes realizados em um dia específico.
* tests_per_million - Número de testes realizados por 1 milhão de pessoas.
* population - Estimativa da população total do país ou território.
* median_age - Idade média da população do país ou território.
* aged_65_older - Percentual da população com 65 anos ou mais.
* aged_70_older - Percentual da população com 70 anos ou mais.
* gdp_per_capita - PIB per capita do país ou território.
* extreme_poverty - Percentual da população vivendo em extrema pobreza.
* cvd_death_rate - Taxa de mortalidade por doenças cardiovasculares.
* diabetes_prevalence - Percentual da população com diabetes.
* female_smokers - Percentual de mulheres adultas que fumam.
* male_smokers - Percentual de homens adultos que fumam.
* handwashing_facilities - Percentual da população com acesso a instalações para *lavar as mãos.
* hospital_beds_per_thousand - Número de leitos hospitalares por mil pessoas.



```
# importar as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')

sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

# importar o dataset COVID-19 para um DataaFrame
data = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/owid-covid-data.csv"
df = pd.read_csv(data)
```

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
from matplotlib.patches import Polygon

x = list(range(168))
z = df_brasil["total_cases"]
z2 = df_brasil["total_deaths"]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))

ax1.set_title("Casos no Brasil")
ax1.set_ylabel("Casos")
ax1.set_xlabel("Dias")
ax1.plot(x, z, color="silver")
ax1.fill_between(x, z, color="red", alpha=0.5)

ax2.set_title("Casos no Brasil (Escala Logarítmica)")
ax2.set_ylabel("Casos")
ax2.set_xlabel("Dias")
ax2.plot(x, z)
ax2.set_yscale("log")
ax2.plot(x, z, color="silver")
ax2.fill_between(x, z, color="red", alpha=0.5)

ax3.set_title("Mortes no Brasil")
ax3.set_ylabel("Mortes")
ax3.set_xlabel("Dias")
ax3.plot(x, z2, color="silver")
ax3.fill_between(x, z2, color="red", alpha=0.5)


ax4.set_title("Mortes no Brasil (Escala logarítmica)")
ax4.set_ylabel("Mortes")
ax4.set_xlabel("Dias")
ax4.plot(x, z2)
ax4.set_yscale("log")
ax4.plot(x, z2, color="silver")
ax4.fill_between(x, z2, color="red", alpha=0.5)

plt.tight_layout()
```

# **Conclusão**

#### *Com base na análise dos dados de COVID-19, podemos destacar alguns pontos chave que ajudam a entender a evolução da pandemia, seus impactos e as variáveis que influenciaram a resposta global.*

1. Crescimento de Casos e Mortes
No início, a pandemia seguiu um crescimento exponencial, com picos de casos e mortes, especialmente durante os primeiros meses. O aumento foi acelerado por fatores como isolamento social, mudanças nas variantes do vírus e a capacidade dos sistemas de saúde de lidar com a carga de pacientes.

2. Diferenças Regionais
Os dados mostram uma grande variação entre países e regiões. Países com respostas rápidas e sistemas de saúde robustos conseguiram controlar melhor a propagação do vírus. Já países com infraestrutura deficiente enfrentaram maiores desafios, refletidos em maiores taxas de mortalidade.

3. Taxa de Letalidade
A análise da taxa de mortalidade indica que, apesar do aumento de casos, muitos países conseguiram reduzir a mortalidade com o tempo, graças a melhorias no tratamento, vacinas e maior conhecimento sobre a doença. No entanto, algumas áreas com maior vulnerabilidade, como populações mais idosas, ainda enfrentaram taxas mais altas.

4. Desafios de Dados
Uma das maiores dificuldades na análise dos dados foi a subnotificação e a inconsistência nas informações entre diferentes fontes. A falta de testes e a variação nos critérios de diagnóstico entre países dificultam uma comparação precisa entre as regiões. Isso é algo a ser considerado ao interpretar os números.

5. Lições para o Futuro
A pandemia de COVID-19 ensinou lições valiosas sobre a importância de sistemas de saúde preparados, a cooperação internacional e a disponibilidade de vacinas e tratamentos para todos. Além disso, a tecnologia e os dados em tempo real serão cruciais para enfrentar futuras crises sanitárias com mais eficiência.

A análise dos dados de COVID-19 destaca os desafios enfrentados pela comunidade científica e a importância de respostas rápidas e bem coordenadas. A pandemia evidenciou a necessidade de sistemas de saúde preparados, além de destacar como o aprendizado contínuo e a adaptação das estratégias de controle, tratamento e prevenção foram essenciais para lidar com a crise. Compreender essas lições é fundamental para estarmos mais bem preparados para futuras emergências de saúde global.
